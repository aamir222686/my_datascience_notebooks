```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('whitegrid')
```

```python
df = sns.load_dataset('iris')
```

```python
map_species = {'virginica': 0, 'versicolor': 1, 'setosa': 2}
```

```python
df['species'] = df['species'].map(map_species)
```

```python
df.head()
```

```python
sns.boxplot(x='species', y='sepal_length', data=df)
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', data=df, hue='species', palette='rainbow')
plt.legend(['one', 'two', 'three'])
plt.legend(loc='lower right')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('species', axis=1)
y = df['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression()
```

```python
log_reg.fit(X_train, y_train)
```

```python
predictions = log_reg.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
print(confusion_matrix(y_test, predictions))
```

```python
round(log_reg.score(X_train, y_train)*100, 2)
```
